# Downloading dataset using accession number

Here, we will see a script to automatically download the raw sequence paired data using project accession number.

```{.sh}
echo "Enter the project identifier (hint: it will be starting from PR...)"
read project

echo "Preparing for the download"

# check if project directory exists
if [ -d "$project" ]; then
  echo "  $project already exists. cd to $project"
else
  echo "  creating a new directory $project"
  mkdir $project
fi

# change to project directory
cd $project
echo "  changing to $project directory"


# check if raw_data directory exists
if [ -d "raw_data" ]; then
  echo "  raw_data already exists. cd to $raw_data"
else
  echo "  creating a new directory raw_data"
  mkdir raw_data
fi

# run search query and store metadata
esearch -db sra -query $project  | efetch -format runinfo > ${project}_metadata.txt


# fetch accessions numbers from metada file into a text file
cat ${project}_metadata.txt | cut -f 1 -d ',' |grep "SRR" > ${project}_sra_id.txt


# change directory to raw_data
cd raw_data
echo "  changing to raw_data directory"

 # download sra file, fastq file and zip fastq file into gz
cat ../${project}_sra_id.txt | while read sra_id; do prefetch $sra_id; fasterq-dump $sra_id; gzip ${sra_id}*.fastq;done

echo "Sequence data has been downloaded successfully"

```

## Accessing abundance data 
Here, we will utilize publicly available processed data of all aforementioned publications. The **dataset is publicly available** and can be downloaded from [here](http://segatalab.cibio.unitn.it/data/Thomas_et_al.html).

Let's load the dataset first and then explore it a bit.

### Loading the dataset


````python
#| echo: true

import pandas as pd
import seaborn as sns
data = pd.read_csv('Nine_CRC_cohorts_taxon_profiles.tsv',sep='\t',header=None).T
data = data.rename(columns=data.loc[0]).drop(0, axis=0)
data.head(5)
````

The processed dataset contains **metadata information** such as `body_site`, `study_condition` along with **abundance data of microbial species**.

We will first start with **Zeller et al., 2014** dataset. For that we will perform filtering based on dataset name and **extract Bacteria kingdom** only.

````python
# accessing Zeller et al., 2014 dataset
zeller_db = data.loc[data['dataset_name'] == 'ZellerG_2014',:]

# fetching microbacterial organism information-related columns
bacteria_colnames = [col for col in data.columns if 'k__Bacteria' in col]

# metadata colnames
metadata_colnames = ['dataset_name', 'sampleID', 'subjectID', 'body_site', 'study_condition',
       'disease', 'age', 'age_category', 'gender', 'country']

# dataframe with bacterial species
zeller_b = zeller_db[bacteria_colnames + ['subjectID']]

#zeller_b.columns = bacteria_colnames + ['subjectID']


# functions to process the data and prepare in a format supporting phyloseq analysis
def get_sample_table(df):
    df = df[metadata_colnames]
    df = df.set_index('subjectID')
    df.drop(['dataset_name','sampleID'],axis=1,inplace=True)
    return df

def get_otu_table(df):
    df = df[bacteria_colnames + ['subjectID']]
    df.columns = ["OTU_{}".format(str(ind)) for ind, col in enumerate(bacteria_colnames)] + ['subjectID']
    new_header = df['subjectID']
    df_new = df.drop(['subjectID'],axis=1)
    df_t = df_new.T
    df_t.columns = new_header
    return df_t

def get_taxa_table(list_of_otus):
    otu = 0
    mapping = {}
    taxa_cols = ['kingdom','phylum','class','order','family','genus','species']
    df = pd.DataFrame(columns=taxa_cols)
    otu_mapping = {}
    otu_ids = []
    for ind, otu in enumerate(list_of_otus):
        tmp = {}
        for col in taxa_cols:
            tmp[col] = get_specific_label(otu, col)
        tmp_df = pd.DataFrame([tmp])
        df = pd.concat([df,tmp_df], ignore_index=[0])

        otu_id = "OTU_{}".format(str(ind))
        otu_mapping[otu] = otu_id
        otu_ids.append(otu)

    df['OTU'] = ['OTU_{}'.format(str(ind)) for ind in df.index]
    df = df.set_index('OTU')
    return df


def get_specific_label(l, t):
    """
    This function parse the taxonomic assignment lable and fetch the specified information (e.g., kingdom, family)

    Args:
        l (str): string of taxonomy
        t (str): string specifying the requested information (e.g., kingdom, family, genus, etc.)

    Returns:
        str: requested heirarichal info 
    """
    taxa_order = {'kingdom':0,'phylum':1,'class':2,'order':3,'family':4,'genus':5,'species':6}

    try:
        specific_label = l.split('|')[taxa_order[t]]

        return specific_label.strip().split('__')[1]
    except:
        return 'Unknown'

# Seperating data into different files
otu_table = get_otu_table(zeller_db)
taxa_table = get_taxa_table(bacteria_colnames)
sample_table = get_sample_table(zeller_db)

otu_table.to_csv('otu_table.csv')
taxa_table.to_csv('taxa_table.csv')
sample_table.to_csv('sample_table.csv')
````

## Visualizing species distribution between healthy and CRC patients
We will now visualize the distribution of species in the dataset among healthy and CRC patients. 

For that we will first aggregate data on family level. The table below shows aggregated abundance at phylum rank.

````python
#| echo: false

for col in otu_table.columns:
    otu_table[col]= otu_table[col].astype('float')

def aggregate_by_taxonomy(otu_table, taxa_table, taxa_rank):
    """
    This function aggregate data based on specified taxa rank (e.g., kingdom, class, order, phylum, genus, species).
    """
    unique_values = (taxa_table[taxa_rank].unique())

    # mapping for otus to unique value of chosen taxa rank
    taxa_to_otu = {}

    # prepare the mapping
    for unique_value in unique_values:
        tdf = taxa_table.loc[taxa_table[taxa_rank] == unique_value, :]
        otus = tdf.index.to_list()   
        taxa_to_otu[unique_value.strip()] = otus
    
    
    # create a dictionary for formulating expressions
    taxa_to_exp = {}

    for key in taxa_to_otu.keys():
        taxa_to_exp[key] = '{} = 0'.format(key)
        for otu in taxa_to_otu[key]:
            taxa_to_exp[key] += ' + ' + otu

    agg_df = otu_table.T

    for key, expr in taxa_to_exp.items():
        agg_df[key] =0
        agg_df = agg_df.eval(expr, engine='python')
        
    agg_df = agg_df[list(unique_values)]
    
    return agg_df

phylum_agg = aggregate_by_taxonomy(otu_table, taxa_table, 'phylum')
genus_agg = aggregate_by_taxonomy(otu_table, taxa_table, 'genus')
order_agg = aggregate_by_taxonomy(otu_table, taxa_table, 'order')
class_agg = aggregate_by_taxonomy(otu_table, taxa_table, 'class')

phylum_df = phylum_agg.merge(sample_table, left_index=True, right_index=True)
genus_df = genus_agg.merge(sample_table, left_index=True, right_index=True)
order_df = order_agg.merge(sample_table, left_index=True, right_index=True)
class_df = class_agg.merge(sample_table, left_index=True, right_index=True)

phylum_df.to_csv('phylum_df.csv')
genus_df.to_csv('genus_df.csv')


phylum_agg.head(5)
````

### Distribution at genus level
At this phase, we filtered out genus with less than **3%** abundance. Then we plotted distribution at genus level for healthy, adenoma and CRC patients. The figures below show distriution of **8** different types of genus in the dataset whose relative abundance satisfied the criteria (>3%).

````python
#| label: fig-charts2
#| fig-cap: "Distribution at genus rank"
#| fig-subcap: 
#|  - Bacteroides
#|  - Prevotella
#|  - Eubacterium
#|  - Blautia
#|  - Coprococcus
#|  - Faecalibacterium
#|  - Ruminococcus
#|  - Subdoligranulum
#| layout-ncol: 3
filter_phylum = taxa_table.loc[taxa_table['phylum'].isin(['Bacteroidetes', 'Firmicutes']),:]

genus_fil = aggregate_by_taxonomy(otu_table,filter_phylum,'genus')

genus_abund = genus_fil.mean(axis=0).to_dict()
genus_list = []

for genus, value in genus_abund.items():
    if value > 3:
        genus_list.append(genus)

genus_new = genus_fil[genus_list]

genus_fil_df = genus_new.merge(sample_table, left_index=True, right_index=True)


for ph in genus_new.columns:
    plt.figure()
    sns.boxplot(x='study_condition', y=ph, data=genus_fil_df, palette=my_pal, saturation=0.2, hue='study_condition')
    plt.title('{} distribution'.format(ph))
    plt.show()
````

### Statistical significance 
Based on our understanding of different taxa at the genus level, we filtered out 8 different types of genus for our statistical analysis. We will use **Kruska Wallis** statistical test for testing the null hypothesis that the **medians of relative abundance of microbial organisms of our selected genus levels are same**. 

There are two different genus (Eubacterium and Ruminococcus) whose differences among healthy, adenoma and CRC patients were found statistically significant.

````python
#| echo: false

from scipy.stats import kruskal
#print(' Genus | p_value ')
for g in genus_list:

    groups = [group[g].values for _, group in  genus_fil_df.groupby("study_condition")]
    stat, p_value = kruskal(*groups)
    #print(f" {g} | {p_value:.3f}")
````

### Alpha diversity
We will now investigate whether their are any differences between healthy, adenoma and CRC patients in terms of alpha diversity.

````python
#| echo: false

from skbio.diversity import alpha

diversity_measures = pd.DataFrame()

alpha_diversity_metrics = [
    "chao1",
    "shannon",
    "simpson",
    "simpson_e",
    "fisher_alpha",
    "berger_parker"
]

shannon_diversity = phylum_agg.apply(lambda x: alpha.shannon(x), axis=1)
chao1_diversity   = phylum_agg.apply(lambda x: alpha.chao1(x), axis=1)
simpson_diversity   = phylum_agg.apply(lambda x: alpha.simpson(x), axis=1)
simpson_e_diversity   = phylum_agg.apply(lambda x: alpha.simpson_e(x), axis=1)
fisher_diversity   = phylum_agg.apply(lambda x: alpha.fisher_alpha(x), axis=1)
berger_parker_diversity   = phylum_agg.apply(lambda x: alpha.berger_parker_d(x), axis=1)

diversity_measures['shannon'] = shannon_diversity
diversity_measures['chao1'] = chao1_diversity
diversity_measures['simpson'] = simpson_diversity
diversity_measures['simpson_e'] = simpson_e_diversity
diversity_measures['fisher_alpha'] = fisher_diversity
diversity_measures['berger_parker'] = berger_parker_diversity

diversity_measures.index = shannon_diversity.index

diversity_df = diversity_measures.merge(sample_table, left_index=True, right_index=True)

diversity_df.to_csv('diversity_metrics.csv')
````

````python
#| label: fig-charts3
#| fig-cap: "Alpha diversity measures"
#| fig-subcap: 
#|  - chao1
#|  - shannon
#|  - simpson
#|  - simpson_e
#|  - fisher_alpha
#|  - berger_parker
#| layout-ncol: 3

for ph in alpha_diversity_metrics:
    plt.figure()
    sns.boxplot(x='study_condition', y=ph, data=diversity_df, palette=my_pal, saturation=0.2, hue='study_condition')
    plt.title('{} metric'.format(ph))
    plt.show()
````
